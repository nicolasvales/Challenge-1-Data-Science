### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)
lista_tiendas = [tienda,tienda2,tienda3,tienda4]
lista_tiendas_nombre= ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
tienda.head()
```

```python
from matplotlib import pyplot as plt
from matplotlib.font_manager import FontProperties
import numpy as np
```

#1. Análisis de facturación



```python
Facturacion_total_tiendas = [sum(tienda['Precio']),sum(tienda2['Precio']),sum(tienda3['Precio']),sum(tienda4['Precio'])]
costo_envio = [sum(tienda['Costo de envío']),sum(tienda2['Costo de envío']),sum(tienda3['Costo de envío']),sum(tienda4['Costo de envío'])]
indice_mayor_venta = Facturacion_total_tiendas.index(max(Facturacion_total_tiendas))
indice_menor_venta = Facturacion_total_tiendas.index(min(Facturacion_total_tiendas))
for i in range(len(lista_tiendas_nombre)):
    print(f'La facturación total de 🏪 {lista_tiendas_nombre[i]} es de 💵 ${Facturacion_total_tiendas[i]}')
    print(f'El gasto total por envios es de 💵 ${costo_envio[i]}\n')
print(f'La 🏪 {lista_tiendas_nombre[indice_mayor_venta]} fue la de mayor facturacion con un monto de: 💵 ${Facturacion_total_tiendas[indice_mayor_venta]}')
print(f'La 🏪 {lista_tiendas_nombre[indice_menor_venta]} fue la de menor facturacion con un monto de: 💵 ${Facturacion_total_tiendas[indice_menor_venta]}')
```

```python
plt.figure(figsize=(10, 6))

bars = plt.bar(
    x=lista_tiendas_nombre,
    height=Facturacion_total_tiendas,
    width=0.6,
    color=['#4CAF50', '#2196F3', '#FF9800', '#800080'],
    edgecolor='black',
    linewidth=1.2
)

plt.title('Facturación total por tienda', fontweight='bold', pad=20)
plt.xlabel('Tiendas', labelpad=10)
plt.ylabel('Facturación', labelpad=10)
plt.xticks(rotation=45, ha='right')

for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height + (0.02 * max(Facturacion_total_tiendas)),
        f'{height:,.0f}',
        ha='center',
        va='bottom',
        fontsize=10
    )

plt.tight_layout()
plt.show()
```

```python
print('De acuerdo a los calculos y al grafico realizado se muestra claramente que la tienda con mayor facturacion es la tienda 1 y la tienda con menor facturacion es la tienda 4')
```

```python
bars = plt.barh(
    y=lista_tiendas_nombre,
    width=costo_envio,
    color=['#4CAF50', '#2196F3', '#FF9800','#800080'],
    edgecolor='black',
    linewidth=1.2
)

plt.title('Gastos de envio por Tienda')
plt.xlabel('Gastos de envio')
plt.ylabel('Tiendas')


for bar in bars:
    width = bar.get_width()
    plt.text(width,
             bar.get_y() + bar.get_height()/2.,
             f'{width:,}',
             ha='left',
             va='center')


plt.tight_layout()
plt.show()
```

```python
print('En le grafico se muestra claramente que la tienda con mayor gasto de envio es la tienda 1 y la tienda con menor gasto de envio es la tienda 4')
```

# 2. Ventas por categoría

```python
def categorias_tiendas(tienda):
    ventas_por_categoria = tienda.groupby('Categoría del Producto')['Categoría del Producto'].count()
    return ventas_por_categoria

for i in range(len(lista_tiendas)):
    nombre_tienda = lista_tiendas_nombre[i]
    cantidad = categorias_tiendas(lista_tiendas[i])
    print(f'🏪 {nombre_tienda}')
    print(f'las ventas por {cantidad}\n')
```

```python
df=pd.DataFrame(columns=lista_tiendas_nombre)
for i in range(len(lista_tiendas)):
    nombre_tienda=lista_tiendas_nombre[i]
    df[nombre_tienda]=categorias_tiendas(lista_tiendas[i])
plt.figure(figsize=(14, 8))
for col in df.columns:
    plt.plot(df.index, df[col], 'o-', label=col, markersize=8)
    for x, y in zip(df.index, df[col]):
        if y > 0:
            plt.text(x, y, f"{int(y)}", ha='center', va='bottom', fontsize=10)

plt.title('Ventas por categoría en todas las tiendas', fontsize=14)
plt.xlabel('Categorías', fontsize=12)
plt.ylabel('Cantidad de ventas', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Tiendas')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda


```python
def Calificación_promedio(tienda):
    promedio=round(sum(tienda['Calificación'])/len(tienda['Calificación']),2)
    return promedio

for i in range (len(lista_tiendas)):
    nombre_tienda=lista_tiendas_nombre[i]
    calificacion=Calificación_promedio(lista_tiendas[i])
    print(f'La 🏪 {nombre_tienda} tiene una ⭐ calificación promedio de: {calificacion}')


```

# 4. Productos más y menos vendidos

```python

def producto_mas_menos_popular(tienda):
    ventas = tienda['Producto'].value_counts()

    producto_mas_popular = ventas.idxmax()
    cantidad_mas_ventas = ventas.max()

    producto_menos_popular = ventas.idxmin()
    cantidad_menos_ventas = ventas.min()

    return producto_mas_popular, cantidad_mas_ventas, producto_menos_popular, cantidad_menos_ventas

for i in range(len(lista_tiendas)):
    tienda_nombre = lista_tiendas_nombre[i]
    mas_popular, cant_mas, menos_popular, cant_menos = producto_mas_menos_popular(lista_tiendas[i])

    print(f'En 🏪 {tienda_nombre}, el 🏷️ producto más popular es {mas_popular} con {cant_mas} ventas')
    print(f'En 🏪 {tienda_nombre}, el 🏷️ producto menos popular es {menos_popular} con {cant_menos} ventas\n')
```

```python
print('De acuerdo el análisis mediante los calculos realizados queda claro con la categoría mas popular en las 4 tiendas es la de muebles y las de menos popularidad es la de Artículos para el hogar en las tiendas 1 y 2;\nY la de Instrumentos musicales en las tiendas 3 y 4.')
```

# 5. Envío promedio por tienda

```python
lista_costo_envio_promedio=[]
for i in range(len(lista_tiendas_nombre)):
  costo_envio_promedio=round(costo_envio[i]/len(lista_tiendas[i]),2)
  lista_costo_envio_promedio.append(costo_envio_promedio)
  print(f'El costo de 🚚 envio promedio de 🏪 {lista_tiendas_nombre[i]} es de ${costo_envio_promedio}')
```

```python
plt.style.use('fivethirtyeight')
group_names=lista_tiendas_nombre
values=lista_costo_envio_promedio
fig, ax = plt.subplots()
ax.barh(group_names, values)
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')
ax.set_xlabel('Costo de envío promedio')
bar_colors = ['#4CAF50', '#2196F3', '#FF9800','#800080']
for bar, color in zip(ax.patches, bar_colors):
    bar.set_facecolor(color)
    bar.set_edgecolor('black')
plt.show()
```

```python
print('El costo total por envios es mayor en la tienda 1 y menor en la tienda 4')
```
